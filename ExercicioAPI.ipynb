```python
import requests
import pandas as pd

# conectar a api e colocar parametros

url = 'https://api.open-meteo.com/v1/forecast'


#pedindo latitute e longitude ao usuario
latitude = float(input('informe a latitude desejada: '))
longitude = float(input('informe a longitude desejada: '))
quantos_dias = int(input('informe quantos dias deseja consultar: '))

# parametros utilizados
parametros = {
    'latitude' : latitude,
    'longitude' : longitude,
    'daily' : ['temperature_2m_max','precipitation_sum'],
    'timezone' : 'auto',
    'past_days' : quantos_dias
}

# como chamar os dados
#pega o os parametros definidos da api e atribiu esses campos na variavel resposta
resposta = requests.get(url, params=parametros)
if resposta.status_code == 200:
  #transformou os dados da resposta da API em formato JSON e atribuiu a variavel dados
  dados = resposta.json()
  # vamos coletar as informações
  #A lista dados_diarios recebeu os dados do parametro daily dentro do get da API
  dados_diarios = dados['daily']

  #separaou os dados para melhor apresentar com o comando print //só para eu lembrar do comando vou usar display que é mais bonito
  #clima = dados_diarios['temperature_2m_max']
  #precipitacao = dados_diarios['precipitation_sum']

  #print(f'Temperatura: {clima} ℃')
  #print(f'Precipitação: {precipitacao} mm')

  #a variavel de dataframe df_dados_clima recebe o conteúdo da lista dados_diarios e esse é tranformado em dataframe pela funcao DataFrame dentro da biblioteca pd
  df_dados_clima = pd.DataFrame(dados_diarios)
  # Transformando em arquivo csv
  df_dados_clima.to_csv('relatorio_prof_luiz.csv', index=False)
  display(df_dados_clima)
else:
    print(f'Erro no sistema numero: {resposta.status_code}')

```
